```python
import os
import requests
from PIL import Image
from io import BytesIO
from tqdm import tqdm
import pandas as pd
from pyinaturalist import get_observations

# Define species list (your provided list)
species_list = [
    # Mammals (25)
    "Lynx lynx", "Ursus arctos", "Canis lupus", "Capreolus capreolus", "Sus scrofa",
    "Martes martes", "Vulpes vulpes", "Lutra lutra", "Rupicapra rupicapra", "Myotis myotis",
    # ... (include all your species)
    "Mantis religiosa"
]

# Create directories
os.makedirs('inaturalist_data', exist_ok=True)
os.makedirs('inaturalist_data/train', exist_ok=True)
os.makedirs('inaturalist_data/val', exist_ok=True)

# Download images for each species
for species in tqdm(species_list):
    # Create species directory
    species_dir = species.replace(" ", "_")
    os.makedirs(f'inaturalist_data/train/{species_dir}', exist_ok=True)
    os.makedirs(f'inaturalist_data/val/{species_dir}', exist_ok=True)

    # Get observations from iNaturalist
    observations = get_observations(
        taxon_name=species,
        quality_grade="research",
        per_page=100  # Get up to 100 images per species
    )

    # Download and save images (80% train, 20% validation)
    for i, obs in enumerate(observations['results']):
        if 'photos' not in obs or not obs['photos']:
            continue

        try:
            img_url = obs['photos'][0]['url'].replace('square', 'medium')  # Get medium size
            response = requests.get(img_url)
            img = Image.open(BytesIO(response.content))

            # Split into train/val
            if i % 5 == 0:  # 20% for validation
                img.save(f'inaturalist_data/val/{species_dir}/{i}.jpg')
            else:
                img.save(f'inaturalist_data/train/{species_dir}/{i}.jpg')
        except Exception as e:
            print(f"Error downloading {species} image {i}: {e}")
```

```python
import torch
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader

# Data augmentation and normalization for training
# Just normalization for validation
data_transforms = {
    'train': transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(20),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}

# Create datasets
image_datasets = {
    'train': datasets.ImageFolder('inaturalist_data/train', data_transforms['train']),
    'val': datasets.ImageFolder('inaturalist_data/val', data_transforms['val'])
}

# Create dataloaders
batch_size = 32
dataloaders = {
    'train': DataLoader(image_datasets['train'], batch_size=batch_size, shuffle=True, num_workers=4),
    'val': DataLoader(image_datasets['val'], batch_size=batch_size, shuffle=False, num_workers=4)
}

# Get class names
class_names = image_datasets['train'].classes
print(f"Classes: {class_names}")
print(f"Training images: {len(image_datasets['train'])}")
print(f"Validation images: {len(image_datasets['val'])}")
```

```python
# Load pre-trained ResNet50
model = models.resnet50(pretrained=True)

# Freeze all layers
for param in model.parameters():
    param.requires_grad = False

# Replace the final fully connected layer
num_classes = len(class_names)
model.fc = torch.nn.Sequential(
    torch.nn.Linear(model.fc.in_features, 512),
    torch.nn.ReLU(),
    torch.nn.Dropout(0.2),
    torch.nn.Linear(512, num_classes)
)

# Move model to GPU if available
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Loss function and optimizer
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.fc.parameters(), lr=0.001)

# Learning rate scheduler
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.1)
```

```python
def train_model(model, criterion, optimizer, scheduler, num_epochs=25):
    best_acc = 0.0

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)

        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Set model to training mode
            else:
                model.eval()   # Set model to evaluate mode

            running_loss = 0.0
            running_corrects = 0

            # Iterate over data
            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device)

                # Zero the parameter gradients
                optimizer.zero_grad()

                # Forward
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    # Backward + optimize only if in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                # Statistics
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            if phase == 'train':
                scheduler.step()

            epoch_loss = running_loss / len(image_datasets[phase])
            epoch_acc = running_corrects.double() / len(image_datasets[phase])

            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

            # Deep copy the model if it's the best so far
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                torch.save(model.state_dict(), 'best_model.pth')

        print()

    print(f'Best val Acc: {best_acc:.4f}')
    return model
```

```python
# Train for 25 epochs
model = train_model(model, criterion, optimizer, scheduler, num_epochs=25)

# Save the final model
torch.save(model.state_dict(), 'final_model.pth')
```

```python
def visualize_predictions(model, num_images=6):
    import matplotlib.pyplot as plt
    import numpy as np

    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure(figsize=(15, 10))

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(dataloaders['val']):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'pred: {class_names[preds[j]]}\ntrue: {class_names[labels[j]]}')

                # Inverse normalization for visualization
                inp = inputs.cpu().data[j].numpy().transpose((1, 2, 0))
                mean = np.array([0.485, 0.456, 0.406])
                std = np.array([0.229, 0.224, 0.225])
                inp = std * inp + mean
                inp = np.clip(inp, 0, 1)

                plt.imshow(inp)

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)

# Visualize some predictions
visualize_predictions(model)
```
